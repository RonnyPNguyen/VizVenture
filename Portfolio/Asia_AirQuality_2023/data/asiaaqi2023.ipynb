```python
import os
import pandas as pd
import json

cwd = os.getcwd()
air = pd.read_csv(cwd + "/asiaaqi2023.csv")
```

```python
air.head()
```

```python
air2= air.copy()
```

```python
for month in ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']:
    air2[month] = pd.to_numeric(air2[month], errors='coerce')
```

```python
air2.isnull().sum()
```

```python
air2 = air2.dropna()
```

```python
air_country = air2.drop(columns=["Rank","City"])
air_country = air_country.groupby(by='Country').mean().sort_values(by='2023', ascending=True).reset_index()
air_country
```

```python
countries = air_country['Country'].unique()
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
# Initialize an empty list for each column
country_list = []
month_list = []
aqi_list = []

for country in countries:
    for month in months:
        country_list.append(country)
        month_list.append(month)
        aqi_list.append(air_country[air_country['Country'] == country][[month]].iloc[0,0])

air_country2 = pd.DataFrame({'Country': country_list, 'Month': month_list, 'AQI': aqi_list}).reset_index(drop=True)
air_country2
```

```python
qualityLabel = air_country2[["Country", "AQI"]].groupby(by="Country").mean().sort_values(by="AQI", ascending = True)
goodLabel = qualityLabel[(qualityLabel["AQI"] <= 12)]
moderateLabel = qualityLabel[(qualityLabel["AQI"] > 12) & (qualityLabel["AQI"] <35.5)]
unhealthyLabel = qualityLabel[(qualityLabel["AQI"] >= 35.5)]

print(goodLabel.shape[0],moderateLabel.shape[0],unhealthyLabel.shape[0])
```

```python
json_data = air_country2.to_json(orient='records', indent=4)

# Export the JSON data to a file
with open('asiaaqi2023.json', 'w') as f:
    f.write(json_data)
```
